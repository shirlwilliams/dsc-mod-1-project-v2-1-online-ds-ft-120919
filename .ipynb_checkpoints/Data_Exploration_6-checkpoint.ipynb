```python
# necessary imports
import os  # for path/directory 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import csv
from importlib import reload
import helper_functions as hf
%matplotlib inline
```

```python
conn = sqlite3.Connection("movies_db.sqlite")
```

```python
# View interested tables
df = hf.load_table(conn, "tmdb_movies")
df.head()
```

### Visualize Movie Ratings versus Popularity

```python
# Scatterplot of movies by popularity vs ratings
sns.scatterplot(x='vote_average', y='popularity', data=df, color='blue')
plt.xlabel('Average IMDB Movie Ratings')
plt.ylabel('Popularity')
plt.title('Movie Ratings versus Popularity')
plt.show()
```

```python
# Table of popularity vs ratings
popularity_vs_rating_df = df[['title', 'popularity', 'vote_average']]
popularity_vs_rating_df.head()
```

```python
top_popularity_vs_rating_df = popularity_vs_rating_df[:100]
```

```python
# Scatterplot of top movies by popularity vs ratings
sns.scatterplot(x='vote_average', y='popularity', data=top_popularity_vs_rating_df, color='red')
plt.xlabel('Average IMDB Voter Rating')
plt.ylabel('Popularity')
plt.title('Top 100 Movie Ratings by Popularity')
plt.show()
```
