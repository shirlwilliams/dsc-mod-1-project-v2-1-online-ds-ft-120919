### Explore Profit by Movie and Studio

```python
query = '''SELECT movie, studio, budget_gross_diff AS profit
            FROM top_budget_gross_diff_movies
            JOIN bom_movie_gross
            ON movie = title
            ORDER BY profit DESC
            '''
profit_by_movie_df = pd.read_sql(query, conn)
profit_by_movie_df.head()
```

```python
profit_by_movie_df = profit_by_movie_df.drop_duplicates().sort_values(by='profit', ascending=False)
profit_by_movie_df.head(10)
```

### Select Profit by Movie Studios

```python
# Create chart for studios by total profit
query = '''SELECT studio, SUM(budget_gross_diff) AS total_profit
                FROM top_budget_gross_diff_movies
                JOIN bom_movie_gross
                ON movie = title
                GROUP BY studio
                ORDER BY total_profit DESC
                            '''
profit_by_studio_df = pd.read_sql(query, conn).head(10)
profit_by_studio_df.head(10)
```

```python
# Format Profit column
pd.set_option('display.float_format', lambda x: '%.0f' % x)
profit_by_studio_df['total_profit'].head()
```

```python
# Bargraph representing the most profitable studios
plt.figure(figsize=(13,8))
sns.set_style('whitegrid')
sns.set(font_scale=1.5)
sns.barplot(x='total_profit', 
            y='studio', 
            data=profit_by_studio_df[:20])
plt.xlabel('Total Profit ( Ten Millions)')
plt.ylabel('Studio')
plt.title('Top 10 Most Profitable Studios', fontsize=20)
```
